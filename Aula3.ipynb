### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
# Analisar clientes
# se o cliente em questão é 'Bom', 'Ok' ou 'Ruim' (Nota de Crédito dele = Score de Crédito)

# Good = Boa
# Standard = Ok
# Poor = Ruim

# Passo a Passo do projeto
# Passo 0: Entender o Desafio da Empresa
# Passo 1: Importar a Base de Dados
# pip install pandas scikit-learn
import pandas as pd

tabela = pd.read_csv("clientes.csv")

display(tabela)
display(tabela.info())
display(tabela.columns)

# int -> números inteiros
# float -> números com casas decimais
# object -> int -> textos
```

```python
# Passo 2:Tratar e Preparar a Base de Dados para a Inteligência artificial

from sklearn.preprocessing import LabelEncoder
import pandas

# Como transformar essas 3 colunas textuais em números?
# profissao, deve ser numérica - algumas profissões: cientista, marceneiro, pedreito, pintor, artista, mecanico

# pode ser feito a partir da seguinte ferramenta:

# LabelEncoder

# ou, pode ser feito manualmente:

# cientista = 1
# marceneiro = 2
# pedreiro = 3
codificador_profissao = LabelEncoder()
tabela["profissao"] = codificador_profissao.fit_transform(tabela["profissao"])

# mix_credito, deve ser numérica
codificador_mix = LabelEncoder()
tabela["mix_credito"] = codificador_mix.fit_transform(tabela["mix_credito"])

# comportamento_pagamento, deve ser numérica
codificador_comportamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])

# Objetivo: prever a coluna "score_credito", somente esta coluna pode ser textual
display(tabela)
```

```python
# Passo 3: Criar o Modelo de Previsão ou Modelo de IA -> 'Ruim', 'Ok', 'Boa'
# escolhendo quais colunas serão usadas para treinar o modelo
# y é a coluna que queremos que o modelo calcule
# x são todas as colunas que serão usadas para prever o'score de crédito', a coluna 'id_cliente' não será usada, porque ela é um número de identificação única, mas como cada linha neste contexto é única, não há necessidade de ter um dado como este: ou seja, ele não ajuda na previsão dos modelos que serão criados.
x = tabela.drop(["score_credito", "id_cliente"], axis=1)
y = tabela["score_credito"]

from sklearn.model_selection import train_test_split

# separar os dados em treino e teste. 'treino', será o nome dado para os modelos aprenderem, enquanto 'teste' será o termo usado para verificar se os modelos aprenderam como se espera.
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# Passo 4: Avaliar e Escolher o Melhor Modelo
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_arvore = RandomForestClassifier()    # modelo árvore de decisão
modelo_knn = KNeighborsClassifier() # modelo do KNN (nearest neighbors - vizinhos mais próximos)

# treinando os modelos
modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

# se o modelo utilizar "Ok" em todos os casos, qual seria a acurácia, a precisão do modelo?
contagem_scores = tabela["score_credito"].value_counts()
print(contagem_scores["Standard"] / sum(contagem_scores))

from sklearn.metrics import accuracy_score

# calcular as previsões
previsao_arvore = modelo_arvore.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

# começar previsões com 'y_teste'
# para esse score o maior é requerido (maior acurácia, mas há a necessidade de o valor ser maior que o fornecido pelo chute de todas as opções como 'Standard')
print(accuracy_score(y_teste, previsao_arvore))
print(accuracy_score(y_teste, previsao_knn))

# quais as características mais importantes para definir o score de crédito?
colunas = list(x_teste.columns)
importancia = pd.DataFrame(index=colunas, data=modelo_arvore.feature_importances_)
importancia = importancia * 100
print(importancia)
```
